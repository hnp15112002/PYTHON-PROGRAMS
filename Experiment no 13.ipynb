Write a Python program to draw line charts of the given data using Matplotlib.

```python
# importing the required libraries
import matplotlib.pyplot as plt
import numpy as np

# define data values
x = np.array([1, 3, 5, 7]) # X-axis points
y = x*3 # Y-axis points

plt.plot(x, y) # Plot the chart
plt.show() # display
```
